```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
import time
import random
```

```python
service = Service(executable_path=r'C:\Users\Виктория\Проект по андану (весна 2024)\chromedriver-win32\chromedriver.exe')
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)

driver = webdriver.Chrome(service=service, options=options)

movies_budgets = []

df = pd.read_csv('df_films_pages_1-13.csv')

first = True
for num, title, link in list(zip(df.index, df['Название'], df['Ссылка'])):
    driver.get(link)
    driver.implicitly_wait(8)
    if first:
        time.sleep(20)
        first = False

    budget, profit_usa, profit_world = None, None, None
    for div in driver.find_elements(by=By.CLASS_NAME, value='styles_rowDark__ucbcz'):
        text = div.text
        if 'Бюджет' in text:
            try:
                budget = text.split('\n')[1]
            except:
                pass
        if 'Сборы в США' in text:
            try:
                profit_usa = text.split('\n')[1]
            except:
                pass
        if 'Сборы в мире' in text:
            try:
                profit_world = text.split('\n')[1].split(' = ')[1]
            except:
                pass
    print(num, '  ', budget, profit_usa, profit_world)
    movies_budgets.append((title, budget, profit_usa, profit_world))
    time.sleep(random.randint(2, 5))


df_budget = pd.DataFrame(movies_budgets, columns=['Название', 'Бюджет($)', 'Сборы в США($)', 'Сборы в мире($)'])



df_budget.to_csv('df_films_budget.csv', index=False)
```
